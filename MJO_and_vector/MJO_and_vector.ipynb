```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from scipy import stats
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 7]
```

```python
def nan_mode(x):
    m = stats.mode(x)[0]
    if not m:
        m = np.NaN
    return m
```

```python
df = pd.read_csv('MJO_Abundance.csv', index_col=0, parse_dates=True)
df['month'] = df.index.month
```

```python
df['Average specimens'].resample('W').mean().plot()
```

```python
# weekly resample
dfs['W'] = df.resample('W').agg({'MJO Phase': nan_mode, 'Average specimens': np.mean})
```

```python
dfs['W']['Average specimens'].plot()
dfs['W']['MJO Phase'].plot()
```

```python
sns.violinplot(x='MJO Phase', y='Average specimens', data=dfs['W'], cut=0)
```
