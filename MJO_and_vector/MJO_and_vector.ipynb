```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 7]

import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('MJO_Abundance.csv', index_col=0, parse_dates=True)
df['month'] = df.index.month
```

```python
df.head()
```

## MJO phase

```python
sns.violinplot(x='MJO Phase', y='Average specimens', data=df, cut=0)
```

## Seasonal

```python
sns.violinplot(x='month', y='Average specimens', data=df, cut=0)
```

## MJO within a season

```python
dfs = dict()
```

```python
Mon = 'Mar'
dfs[Mon] = df[df['month']==3]
sns.violinplot(x='MJO Phase', y='Average specimens', data=dfs[Mon], cut=0)
```

```python
Mon = 'May'
dfs[Mon] = df[df['month']==5]
sns.violinplot(x='MJO Phase', y='Average specimens', data=dfs[Mon], cut=0)
```

```python
Mon = 'Apr'
dfs[Mon] = df[df['month']==4]
sns.violinplot(x='MJO Phase', y='Average specimens', data=dfs[Mon], cut=0)
```
